## __What should we learn?__

<div class="alert alert-block alert-info">
<b>Basic Understanding</b>
    <pre>
1.  <b>Linear Algebra </b>
    Basic of vectors, matrices and matrix operations, vector norms, Euclidean distance, Manhattan distance. 
2.  <b>Statistics </b>
    Descriptive statistics, correlation and covariance, methods of measurement of correlations.
3.  <b>Probability</b>
    Basics of probability theory, joint/conditional/marginal probability, Bayes' theorem.
4.  <b>Calculus</b>
    Fundamental Calculus Concepts , gradient, Partial Derivatives, Higher-Order Derivatives.
    <pre>
</div>

<div class="alert alert-block alert-warning">
<b>Intermediate Understanding</b>
    <pre>
1.  <b>Linear Algebra </b>
    Eigenvalues and Eigenvectors, LU Decomposition, Singular Value Decomposition (SVD).
2.  <b>Statistics dan Probability </b>
    Central Limit Theorem, Discrete Probability Distributions, Continuous Probability Distributions, hypothesis testing and confidence intervals.
3.  <b>Calculus</b>
    Partial Derivatives and chain rule for backpropagation in neural networks.
4.  <b>Optimization</b>
    Understanding gradient descent and its variations (e.g., stochastic gradient descent).
    <pre>
</div>

<div class="alert alert-block alert-danger">
<b>Advanced Understanding</b>
    <pre>
1.  <b>Vector Calculus </b>
    Jacobian, Hessian Matrices.
2.  <b>Probability Distributions and Statistics </b>
    Sampling Distributions, Chi-Square Distribution, t -Distribution, Parametric Methods, Non-Parametric Test, Bias Vs Variance and Bootstrap method.
3.  <b>Geometry</b>
    Cosine Similarity, Jaccard Similarity and Orthogonality and Projections.
4.  <b>Regression Analysis </b>
    Maximum Likelihood Estimation (MLE), Mean Squared Error.
    <pre>
</div>
